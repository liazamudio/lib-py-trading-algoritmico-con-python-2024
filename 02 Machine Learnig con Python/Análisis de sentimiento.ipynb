# Analisis de sentimiento

```python
import requests
from textblob import TextBlob

# Configurar la URL y los parámetros de la API de News API
url = "https://paper-api.alpaca.markets/v2"
api_key = "PKHTH7QHJ2MDU6XJUWO0"

# Parámetros de búsqueda
query = "Apple"
language = "es"

# Configurar los parámetros de la solicitud
params = {
    "q": query,
    "language": language,
    "apiKey": api_key
}

# Realizar la solicitud GET a la API de News API
response = requests.get(url, params=params)
data = response.json()

# Analizar el sentimiento de cada noticia
for article in data["articles"]:
    title = article["title"]
    description = article["description"]

    # Concatenar el título y la descripción para el análisis de sentimiento
    text = f"{title} {description}"

    # Crear una instancia de TextBlob
    blob = TextBlob(text)

    # Obtener la polaridad del sentimiento
    polarity = blob.sentiment.polarity

    # Obtener la subjetividad del sentimiento
    subjectivity = blob.sentiment.subjectivity

    # Imprimir los resultados
    print("Noticia:")
    print("Título:", title)
    print("Descripción:", description)
    print("Sentimiento:")
    print("Polaridad:", polarity)
    print("Subjetividad:", subjectivity)
    print("-----------------------------------------")
```

# Sentimiento Medio

```python
import requests
from textblob import TextBlob

# Configurar la URL y los parámetros de la API de News API
url = "https://paper-api.alpaca.markets/v2"
api_key = "PKHTH7QHJ2MDU6XJUWO0"

# Parámetros de búsqueda
query = "empresa"
language = "es"

# Configurar los parámetros de la solicitud
params = {
    "q": query,
    "language": language,
    "apiKey": api_key
}

# Realizar la solicitud GET a la API de News API
response = requests.get(url, params=params)
data = response.json()

# Variables para calcular la media
total_polaridad = 0
total_subjetividad = 0
num_noticias = 0

# Analizar el sentimiento de cada noticia
for article in data["articles"]:
    title = article["title"]
    description = article["description"]

    # Concatenar el título y la descripción para el análisis de sentimiento
    text = f"{title} {description}"

    # Crear una instancia de TextBlob
    blob = TextBlob(text)

    # Obtener la polaridad del sentimiento
    polaridad = blob.sentiment.polarity

    # Obtener la subjetividad del sentimiento
    subjetividad = blob.sentiment.subjectivity

    # Acumular los valores
    total_polaridad += polaridad
    total_subjetividad += subjetividad
    num_noticias += 1

    # Imprimir los resultados
    print("Noticia:")
    print("Título:", title)
    print("Descripción:", description)
    print("Sentimiento:")
    print("Polaridad:", polaridad)
    print("Subjetividad:", subjetividad)
    print("-----------------------------------------")

# Calcular la media
media_polaridad = total_polaridad / num_noticias
media_subjetividad = total_subjetividad / num_noticias

# Imprimir los resultados de la media
print("Media:")
print("Media de polaridad:", media_polaridad)
print("Media de subjetividad:", media_subjetividad)
```
