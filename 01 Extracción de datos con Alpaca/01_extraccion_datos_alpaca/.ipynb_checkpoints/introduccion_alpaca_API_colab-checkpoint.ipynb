# Uso de una Estrategia basada en RSI

El Indicador de Fuerza Relativa, o RSI por sus siglas en inglés, es un indicador técnico utilizado en el análisis técnico para medir la fuerza de la tendencia y detectar posibles cambios en la dirección de la misma.

Para calcular el RSI, se utiliza la siguiente fórmula:

RSI = 100 - (100 / (1 + RS))

Donde RS (Relative Strength) se calcula dividiendo la media de los precios de cierre en días alcistas entre la media de los precios de cierre en días bajistas durante un periodo de tiempo determinado.

Una vez que se ha calculado el valor del RSI, se pueden establecer los niveles de sobrecompra y sobreventa en 70 y 30, respectivamente. Cuando el RSI sube por encima de 70, se considera que el activo está sobrecomprado y se espera una posible corrección a la baja en el precio. Por otro lado, cuando el RSI cae por debajo de 30, se considera que el activo está sobrevendido y se espera una posible corrección al alza en el precio.

```python
import ta
import pandas as pd

# Calcular RSI de 14 periodos
df['rsi'] = ta.momentum.RSIIndicator(df['close'], window=14).rsi()
```

```python
df['Buy'] = df['rsi'].lt(30).astype(int)
df['Sell'] = df['rsi'].gt(70).astype(int)

# Mostrar DataFrame con señales
print(df)
```

```python
df.head(20)
```

```python
import numpy as np
df['signal'] = np.where(df['rsi']> 70, 'SELL', np.where(df['rsi']< 30, 'BUY', ''))
```

# Ejecución de Ordenes

```python
position = 0
for i in range(len(df)):
    if df['signal'][i] == 'BUY':
        if position == 0:
            api.submit_order(
                symbol=symbol,
                qty=1,
                side='buy',
                type='market',
                time_in_force='gtc'
            )
            position = 1
    elif df['signal'][i] == 'SELL':
        if position == 1:
            api.submit_order(
                symbol=symbol,
                qty=1,
                side='sell',
                type='market',
                time_in_force='gtc'
            )
            position = 0
```

```python
import datetime
daily_rsi = df['rsi'] = ta.momentum.RSIIndicator(df['close'], window=14).rsi()

# Obtener la fecha actual
current_date = datetime.date.today()

# Verificar que la fecha actual coincide con la última fecha en el DataFrame
last_date = df['time'].index[-1]
if current_date == last_date:
    # Obtener el RSI diario más reciente
    current_rsi = daily_rsi.iloc[-1]['rsi']

    # Comprobar las señales de compra y venta según los niveles de sobrecompra y sobreventa de 70 y 30, respectivamente.
    if current_rsi < 30:
        # Generar una orden de compra
        api.submit_order(
            symbol=symbol,
            qty=1,
            side='buy',
            type='market',
            time_in_force='gtc'
        )
    elif current_rsi > 70:
        # Generar una orden de venta
        api.submit_order(
            symbol=symbol,
            qty=1,
            side='sell',
            type='market',
            time_in_force='gtc'
        )
    else:
        print('No se generó ninguna orden.')
else:
    print('La fecha actual no coincide con la última fecha en el DataFrame.')
```
